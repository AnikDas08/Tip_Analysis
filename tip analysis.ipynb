```python
#import data set
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style("whitegrid")
```

```python
#load dataset
df=pd.read_csv("tips.csv")
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df["total_bill"].sum()
```

```python
lowest_tip=df["tip"].min()
higest_tip=df["tip"].max()
avg_tip=df["tip"].mean()
print(f"Higest_tip : {higest_tip}\n lowest_tip : {lowest_tip}\navg_tip : {avg_tip}")
```

```python
bill_sum=df["total_bill"].sum()
tip_sum=df["tip"].sum()
percent=tip_sum/bill_sum*100
print(f"Avarage tip is {percent}% of bill")
```

```python
corrMatrix=df.corr(numeric_only=True)
sns.heatmap(corrMatrix,annot=True)
plt.show()
```
